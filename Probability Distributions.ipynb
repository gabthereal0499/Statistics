```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint
```

#Uniform Distribution

A uniform distribution has constant probability with in a given range

The uniform distribution is a continuous distribution in which all outcome in an interval [a,b]are equally likely




it is defined by parameters:

*   Lower bounnd

*   Upper bound

eg:Modeling the probability of a arandom point along line statement





```python
from scipy.stats import uniform

#parameter
a,b=0,10
x_values=np.linspace(a,b,1000)

#plot PDF

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(x_values,uniform.pdf(x_values,a,b-a),color="blue")
plt.title("uniform Distribution PDF (a=0,b=10)")
plt.xlabel("x")
plt.ylabel("PDF")


#plt CDF
plt.subplot(1,2,2)
plt.plot(x_values,uniform.cdf(x_values,a,b-a),color="green")
plt.title("uniform Distribution CDF (a=0,b=10)")
plt.xlabel("x")
plt.ylabel("CDF")
```

```python
#Parameteers for discrete Uniform Distribution
low,high=1,6 #inclusive range from 1 to 6 (like a fair die)
x_values=np.arange(low,high+1)

plt.figure(figsize=(12,6))
#plot PMT
plt.subplot(1,2,1)
plt.stem(x_values,randint.pmf(x_values,low,high+1),basefmt="")
plt.title("Discrete Uniform Distribution PMF(low=1,high=6)")
plt.xlabel("x")
plt.ylabel("PMF")

#plot CDF
plt.subplot(1,2,2)
plt.step(x_values,randint.cdf(x_values,low,high+1),color="green")
plt.title("Discrete Uniform Distribution CDF (low=1,high=6)")
plt.xlabel("x")
plt.ylabel("CDF")
plt.show()
```

#Bernoulli Distribution



The bernoulli Distribution models a single experiment with two possible outcomes:success (with probability ~p ) and failure (with probability 1-p)


it is a discrete distribution and defined by the parameter
- p :probability of success

Example: Modelling a single coin flip where heads is a success and tails is a failure

```python

from scipy.stats import bernoulli

#parameters
p=0.5
x_values=np.array([0,1])

plt.figure(figsize=(12,6))

#plot PMF

plt.subplot(1,2,1)
plt.stem(x_values,bernoulli.pmf(x_values,p),basefmt="")
plt.title("Bernoulli Distribution PMF (p=0.5)")
plt.xlabel("x")
plt.ylabel("PMF")
```

#Normal Distribution

The Normal Disrtibution also known as Gaussian distribution ,is a continous distribution defined by its symmetric,bell-shaped curve centered around the mean.
it's defined by parameters:


*   "mu":Mean of the distribution
*   "sigma":standard deviation,which ditermines the spred of the distribution

Special case:The standard Normal Distribution tha "mu=0" and "sigma=1"

eg:modeling height,test score or measurement errors



```python
from scipy.stats import norm
#parameters for normal distribution
mu=1 #mean
sigma=3 #standerd deviation
x=np.linspace(-10,10,100)

plt.figure(figsize=(12,6))
#plot PDF
plt.subplot(1,2,1)
plt.plot(x,norm.pdf(x,mu,sigma),color="blue",label="PDF")
plt.title("Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend

#CDF
plt.subplot(1,2,2)
plt.plot(x,norm.cdf(x,mu,sigma),color="red",label="CDF")
plt.title("Normal Distribution CDF")
plt.xlabel("x")
plt.ylabel("p(xx)")
plt.grid(True)
plt.legend()
plt.show()
```
