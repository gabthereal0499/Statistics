```python
#importing the necessary libraries
import numpy as np
import statistics as stats
```

```python
#data for eg:
data=[10,15,20,25,30,35,40,45,50,55,60]
```

```python
#mean using numpy
mean_value=np.mean(data)
print("mean is:",mean_value)
```

```python
mean_value=stats.mean(data)
print("mean is:",mean_value)
```

```python
print(stats.mean([10,15,20,25,30,35,40,45,50,55,60]))
```

#MEDIAN


# New Section

```python
#median example
median_value=np.median(data)
print("median is:",median_value)
```

if the number of data values is odd,it returns the exact middle value.if the no of data value is even, it returns the average of the two middle value



```python
print(stats.median([1,3,5,7,9,11,13]))
#the number of data value is odd
```

```python
print(stats.median([1,3,5,7,9,11]))
#the no of data value is even
```

```python
print(stats.median([-11,5.5,-3.4,7.1,-9,22]))
#-11,-9,-3.4,5.5,7.1,22  after sorting
```

##median high

calculate the high median (middle value) of the given data.if the number of data value is even,it returns the larger of two middle value .

```python
print(stats.median_high([1,3,5,7,9,11]))
```

##median low

```python
print(stats.median_low([1,3,5,7,9,11]))
```

```python
print(stats.median_low([-11,5.5,-3.4,7.1,-9,22]))
```

# mode

```python
mode_value=stats.mode(data)
print("mode is:",mode_value)
```

if all value occure with the sama frequency,statics.mode()return the first value in order of the data set, which is 10 this case

```python
data_with_mode=[1,2,2,2,3,3,3,4,4,4,4,4,4,5]
mode_value=stats.mode(data_with_mode)
print(mode_value)
```

## quanditative data
